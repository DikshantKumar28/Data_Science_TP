# API Handling

```python
import requests
import pandas as pd

# API URL
url = "https://jsonplaceholder.typicode.com/users"

# Send request
response = requests.get(url)

# Convert JSON
data = response.json()

# Empty list
users_data = []

# Extract specific fields
for user in data:

    user_obj = {

        "ID": user["id"],

        "Name": user["name"],

        "Username": user["username"],

        "Email": user["email"],

        "City": user["address"]["city"],

        "Latitude": float(
            user["address"]["geo"]["lat"]
        )
    }

    users_data.append(user_obj)

# Create DataFrame
df = pd.DataFrame(users_data)

# ==========================
# DATAFRAME OPERATIONS
# ==========================
print(df.head())


print(
    df["Latitude"].mean()
)

print(
    df.sort_values(
        by="Latitude",
        ascending=True
    )
)


print(
    df.sort_values(
        by="Latitude",
        ascending=False
    )
)
print(
    df["Latitude"].std()
)

print(df.info())

print(df.describe())
```
