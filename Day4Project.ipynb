# Sales Performance Dashboard(Data Layer)

#### Importing library and dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("Superstore.csv", encoding="latin-1")
```

#### Understanding GroupBy & Finding Total Sales by Region


```python
df.head()
df.groupby("Region")["Sales"].sum()

```

#### Multiple Grouping

```python
df.groupby(["Region", "Category"])[["Sales", "Profit"]].sum()
```

#### Understanding Aggregation

```python
df.groupby("Region").agg({
    "Sales": ["sum", "mean", "max"],
    "Profit": ["sum", "mean"]
})
```

#### Top 3 Sub-Categorries Per Region

```python
subcat_sales = df.groupby(
    ["Region", "Sub-Category"]
)["Sales"].sum()
```

```python
subcat_sales = subcat_sales.sort_values(ascending=False)
```

```python
top3 = subcat_sales.groupby(level=0).head(3)
top3
```

#### Add New Column

```python
df["Profit Margin %"] = (
    df["Profit"] / df["Sales"]
) * 100
```

#### Pivot Tables

```python
pivot = pd.pivot_table(
    df,
    values="Sales",
    index="Region",
    columns="Segment",
    aggfunc="mean"
)
```

#### Loss making month calculations

```python
# Datetime conversion
df["Order Date"] = pd.to_datetime(
    df["Order Date"]
)

# Monthly profit
df["Month"] = df["Order Date"].dt.month

monthly_profit = df.groupby(
    "Month"
)["Profit"].sum()

loss_months = monthly_profit[
    monthly_profit < 0
]

print(loss_months)

# Quarterly sales
df = df.set_index("Order Date")

quarterly_sales = df["Sales"].resample(
    "Q"
).sum()

print(quarterly_sales)
```
