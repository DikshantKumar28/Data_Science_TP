# Database Handling using SQLite 

```python
import sqlite3

# Create/connect database
conn = sqlite3.connect("company_master.db")

print("Database Connected")
```

```python
cursor = conn.cursor()
```

```python
import pandas as pd
employees_df=pd.read_sql_query("SELECT * FROM employees", conn)
```

```python
employees_df.head()
```

```python
conn2=sqlite3.connect("project_records.db")
```

```python
projects_df = pd.read_sql_query(
    "SELECT * FROM projects",
    conn2
)

print(projects_df.head())
```

```python
merged_df=pd.merge(employees_df, projects_df, on="employee_id", how="inner")
```
