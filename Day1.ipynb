##### write a program which cotnains different functions
##### converts list to matrix format of 3*3
 ##### convert the same list into a tuple and find the mean, median,mode
  ##### third function will iterate over this two functions and find the gaps, fill with null values respectively and gives a dicitionary format, mapped with first to the second function

```python
import statistics as st


# Original List
lst = [1, 2, 3, 4, 5, 6, 7, 8, 9]




def convert_matrix(data):

    mat= []

    for i in range(0, len(data), 3):
        row = data[i:i+3]
        mat.append(row)

    return mat



def calculation_score(data):

    tup = tuple(data)

    tup_mean = st.mean(tup)
    tup_med = st.median(tup)
    tup_mod = st.mode(tup)

    return {
        "Tuple": tup,
        "Mean": tup_mean,
        "Median": tup_med,
        "Mode": tup_mod
    }




def map_functions(data):

    matrix_output = convert_matrix(data)

    stats_output = calculation_score(data)

    result = {}

    max_length = max(len(matrix_output), len(stats_output))

    matrix_keys = list(range(len(matrix_output)))
    stats_keys = list(stats_output.keys())

    for i in range(max_length):

        key = matrix_keys[i] if i < len(matrix_keys) else None

        value = stats_keys[i] if i < len(stats_keys) else None

        result[key] = value

    return result




print("3x3 Matrix:")
print(convert_matrix(lst))

print("\nTuple Statistics:")
print(calculation_score(lst))

print("\nMapped Dictionary:")
print(map_functions(lst))
```
