# Student Grade Analyzer- NumPy Project

```python
import numpy as np
```

```python
marks=np.random.randint(13,100,size=50)
marks
```

#### Basics Statistics 

```python
mean_score=np.mean(marks)
median_score=np.median(marks)
highest_score=np.max(marks)
lowest_score=np.min(marks)
std_dev_score=np.std(marks)
```

#### Find Distinction Students

```python
## Failed students first
failed=marks[marks<40]
print(f"Failed students:{failed}")
```

```python
### Students scoring above 85
top_scorers=marks[marks>85]
print(f"Top scorers:{top_scorers}")
```

#### Normalize Scores

```python
normalized_scores = (marks - np.min(marks)) / (np.max(marks) - np.min(marks))

print("Normalized Scores:")
print(normalized_scores)
```

#### Reshaping the array


```python
reshaped_marks = marks.reshape(5, 10)

print(reshaped_marks)
```

#### Row-wise Average

```python
subject_averages = np.mean(reshaped_marks, axis=1)

print("Subject-wise Averages:")
print(subject_averages)
```

#### Percentiles

```python
p25 = np.percentile(marks, 25)
p50 = np.percentile(marks, 50)
p90 = np.percentile(marks, 90)

print("25th Percentile:", p25)
print("50th Percentile:", p50)
print("90th Percentile:", p90)
```

#### Bonus Challenge

```python
theory = np.random.randint(0, 100, 50)
practical = np.random.randint(0, 100, 50)

final_scores = (0.4 * theory) + (0.6 * practical)

print("Theory Scores:")
print(theory)

print("Practical Scores:")
print(practical)

print("Weighted Final Scores:")
print(final_scores)
```
