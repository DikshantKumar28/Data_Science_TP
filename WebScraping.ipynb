# Web Scraping

```python
import requests as req 
from bs4 import BeautifulSoup as bs
import pandas as pd
import re 
```

```python
url="http://books.toscrape.com/"
response=req.get(url)
```

```python
soup=bs(response.text,"html.parser")
```

```python
books = soup.find_all("article", class_="product_pod")
```

```python
titles = []
prices = []
ratings = []
```

```python
for book in books: 
    title = book.h3.a["title"]
    
    price = book.find(
        "p",
        class_="price_color"
    ).text
    
    rating = book.p["class"][1]
    
    titles.append(title)
    prices.append(price)
    ratings.append(rating)
```

```python
clean_prices = []   
for p in prices: 
     # Remove everything except numbers and decimal
    value = float(
        re.sub(r"[^\d.]", "", p)
    )

    clean_prices.append(value)
```

```python
# Create DataFrame
df = pd.DataFrame({
    "Title": titles,
    "Price": clean_prices,
    "Rating": ratings
})
```

```python
# 1. Display first 5 rows

print(df.head())

# 2. Mean of Price column


mean_price = df["Price"].mean()

print(mean_price)

```

```python
# 3. Sort Ascending

ascending_df = df.sort_values(
    by="Price",
    ascending=True
)

print(ascending_df)

```

```python
# 4. Sort Descending

descending_df = df.sort_values(
    by="Price",
    ascending=False
)

print(descending_df)
```

```python

# 5. Standard Deviation


std_price = df["Price"].std()

print(std_price)
```
